Segmenting and Clustering Neighborhoods in the city of Toronto, Canada

Web Scapping using BeautifulSoup and creating data frame using pandas

```python
import pandas as pd

import requests

from bs4 import BeautifulSoup

req = requests.get("https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&oldid=942851379")

soup = BeautifulSoup(req.content,'lxml')

table = soup.find_all('table')[0]

df = pd.read_html(str(table))

neighborhood=pd.DataFrame(df[0])
neighborhood
```

Ignore cells with a borough that is Not assigned.

```python
neighborhood=neighborhood[neighborhood.Borough != 'Not assigned']
neighborhood
```

1. Check if a cell has a borough but a Not assigned neighborhood, then replace that neighborhood with borough
2. Find duplicate Post Codes and merge its neighborhoods in a comma separated format.

```python
neighborhood.loc[(neighborhood['Borough'] != 'Not assigned') & (neighborhood['Neighbourhood'] == 'Not assigned'), 'Neighbourhood'] = neighborhood['Borough']
neighborhood
neighborhood=neighborhood.groupby(['Postcode','Borough']).Neighbourhood.agg([('Neighbourhood', ', '.join)]).reset_index()
neighborhood
```

```python
locations = pd.read_csv("https://cocl.us/Geospatial_data")
locations.head()
```

```python
locations.columns = ["Postal Code", "Latitude", "Longitude"]
locations.rename(columns={'Postal Code': 'Postcode'}, inplace=True)
neighborhood2 = pd.merge(neighborhood, locations, on='Postcode')
neighborhood2.head()
```

```python
neighborhood2[neighborhood2["Postcode"] == "M5G"]
```

```python
address = 'New York City, NY'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of New York City are {}, {}.'.format(latitude, longitude))
```

```python
# create map of New York using latitude and longitude values
map_newyork = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(neighborhood2['Latitude'], neighborhood2['Longitude'], neighborhood2['Borough'], neighborhood2['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_newyork)  
    
map_newyork
```
